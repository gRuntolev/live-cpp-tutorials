# State persistence across cells in xeus-cpp



#### In a standard C++ development invironment, each time you run a program, the memory is allocated, the `main()` function executes, and then the memory is completely wiped. In **xeus-cpp**, the execution model is **persistent**

---


### What is state persistence?

The Jupyter notebook acts as a single, ongoing C++ session. When you execute a cell, the kernel compiles that specific snippet and integrates it into the current process memory. This means, that any variable, function or object defined in a cell remains alive and accessible in any other cell.

***NB! The order in which you run cells matters. Running cells out of order can cause errors and unexpected behaviour***


### The following program demonstrates how state persistence in xeus-cpp works exactly:

```c++
int counter = 0;
```

```c++
#include <iostream>

void inc()
{
    counter++;
    std::cout << "Current Value: " << counter << std::endl;
}
```

```c++
inc(); //Run this cell multiple times to see persistence in action
```
