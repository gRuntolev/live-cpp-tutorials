## 1. Function Templates

```c++
// swap
template<typename T>
void mySwap(T& a, T& b) {
    T temp = move(a);
    a = move(b);
    b = move(temp);
}

// max
template<typename T>
const T& myMax(const T& a, const T& b) {
    return (a > b) ? a : b;
}

// printAll
template<typename T>
void printAll(const vector<T>& v) {
    for (const T& x : v) cout << x << " ";
    cout << endl;
}

int x = 10, y = 20;
mySwap(x, y);
cout << "After swap: x = " << x << " y = " << y << endl;
cout << "max(3.14, 2.72) = " << myMax(3.14, 2.72) << endl;

vector<string> words = {"Plovdiv", "Sofia", "Varna"};
printAll(words);
```

## 2. Class Templates — Generic Stack

```c++
template<typename T, size_t MaxSize = 100>
class Stack {
    vector<T> data;
public:
    void push(const T& val) {
        if (data.size() >= MaxSize) throw overflow_error("Stack is full");
        data.push_back(val);
    }

    T pop() {
        if (empty()) throw underflow_error("Stack is empty");
        T val = move(data.back());
        data.pop_back();
        return val;
    }

    const T& top() const {
        if (empty()) throw underflow_error("Stack is empty");
        return data.back();
    }

    bool empty() const { return data.empty(); }
    size_t size() const { return data.size(); }
};

// Use with different types
Stack<int> intStack;
for (int i : {1, 5, 3, 8, 2}) intStack.push(i);
cout << "Popping int stack: ";
while (!intStack.empty()) cout << intStack.pop() << " ";
cout << endl;

Stack<string> strStack;
for (const string& s : {"hello"s, "world"s, "!"s}) strStack.push(s);
cout << "Popping string stack: ";
while (!strStack.empty()) cout << strStack.pop() << " ";
cout << endl;
```

## 3. Template Specialization

```c++
// Generic printer
template<typename T>
struct Printer {
    static void print(const T& val) { cout << "Value: " << val << endl; }
};

// Specialization for bool
template<>
struct Printer<bool> {
    static void print(const bool& val) {
        cout << "Bool: " << (val ? "true" : "false") << endl;
    }
};

// Specialization for vector<T>
template<typename T>
struct Printer<vector<T>> {
    static void print(const vector<T>& v) {
        cout << "Vector[" << v.size() << "]: ";
        for (const T& x : v) cout << x << " ";
        cout << endl;
    }
};

Printer<int>::print(42);
Printer<bool>::print(true);
Printer<vector<int>>::print({1, 2, 3, 4, 5});
```

## 4. Variadic Templates

```c++
// Sum of any number of arguments
template<typename T>
T mySum(T val) { return val; }

template<typename T, typename... Args>
T mySum(T first, Args... rest) { return first + mySum(rest...); }

cout << "Sum(1,2,3) = " << mySum(1, 2, 3) << endl;
cout << "Sum(1.1,2.2,3.3) = " << mySum(1.1, 2.2, 3.3) << endl;
cout << "Sum(1,2,3,4,5) = " << mySum(1, 2, 3, 4, 5) << endl;

// Variadic print
template<typename... Args>
void printMany(Args... args) {
    ((cout << args << " "), ...) ;
    cout << endl;
}
printMany("hello", 42, 3.14, true, 'X');
```

## 5. Multiple Inheritance

```c++

class Serializable {
public:
    virtual string serialize() const = 0;
    virtual ~Serializable() = default;
};

class Printable {
public:
    virtual void print() const = 0;
    virtual ~Printable() = default;
};

class Person : public Serializable, public Printable {
    string name;
    int age;
public:
    Person(const string& n, int a) : name(n), age(a) {}
    string serialize() const override {
        return "{name:\"" + name + "\",age:" + to_string(age) + "}";
    }
    void print() const override {
        cout << "Person: " << name << ", age " << age << endl;
    }
};

Person p("Andrew", 20);
p.print();
cout << "Serialized: " << p.serialize() << endl;
```
