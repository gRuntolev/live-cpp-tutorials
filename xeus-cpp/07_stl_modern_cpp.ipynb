## 1. std::vector

```c++
vector<int> v = {5, 2, 8, 1, 9, 3, 7, 4, 6};

// Sort and display
sort(v.begin(), v.end());
cout << "Sorted: ";
for (int x : v) cout << x << " ";
cout << endl;

// Accumulate
int sum = accumulate(v.begin(), v.end(), 0);
cout << "Sum: " << sum << "  Average: " << (double)sum / v.size() << endl;

// Find
auto it = find(v.begin(), v.end(), 7);
cout << "Found 7 at index: " << distance(v.begin(), it) << endl;
```

```c++
// Transform: square each element
vector<int> squares(v.size());
transform(v.begin(), v.end(), squares.begin(), [](int x) { return x * x; });
cout << "Squares: ";
for (int x : squares) cout << x << " ";
cout << endl;
```

## 2. std::map and std::set

```c++
// Word frequency counter
string text = "the quick brown fox jumps over the lazy dog";
map<string, int> freq;

string word;
stringstream ss(text);
while (ss >> word) freq[word]++;

cout << "Word frequencies (alphabetical):" << endl;
for (const auto& [w, cnt] : freq) {
    cout << "  " << w << ": " << cnt << endl;
}
```

```c++
// Set: unique sorted elements
vector<int> dupes = {3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5};
set<int> unique_set(dupes.begin(), dupes.end());

cout << "Original (" << dupes.size() << " elements): ";
for (int x : dupes) cout << x << " ";
cout << "\nUnique (" << unique_set.size() << " elements): ";
for (int x : unique_set) cout << x << " ";
cout << endl;
```

## 3. Lambda Expressions

```c++
// Lambda captures
[&]() {
int threshold = 5;
auto isAbove = [threshold](int x) { return x > threshold; };  // capture by value
auto addTo   = [&threshold](int x) { return x + threshold; };  // capture by reference

vector<int> nums = {1, 3, 5, 7, 9, 2, 4, 6, 8};
vector<int> above, shifted;

copy_if(nums.begin(), nums.end(), back_inserter(above), isAbove);
transform(nums.begin(), nums.end(), back_inserter(shifted), addTo);

cout << "Above " << threshold << ": ";
for (int x : above) cout << x << " ";
cout << "\nShifted +" << threshold << ": ";
for (int x : shifted) cout << x << " ";
cout << endl;
}();
```

## 4. Smart Pointers

```c++
// unique_ptr — single ownership
{
    auto uptr = make_unique<string>("unique owner");
    cout << "unique_ptr: " << *uptr << endl;
    // auto uptr2 = uptr;  // ERROR: cannot copy unique_ptr
    auto uptr2 = move(uptr);  // Transfer ownership
    cout << "After move, uptr is " << (uptr ? "valid" : "null") << endl;
    cout << "uptr2: " << *uptr2 << endl;
}  // uptr2 automatically freed here

// shared_ptr — shared ownership
{
    auto sptr1 = make_shared<int>(42);
    auto sptr2 = sptr1;  // Both share ownership
    cout << "shared value: " << *sptr1 << "  use_count: " << sptr1.use_count() << endl;
}  // Both freed here
```

## 5. C++17 Features: optional and variant

```c++
// std::optional
optional<int> safeDivide(int a, int b) {
    if (b == 0) return nullopt;
    return a / b;
}

auto r1 = safeDivide(10, 2);
auto r2 = safeDivide(10, 0);

cout << "10/2 = " << (r1 ? to_string(*r1) : "undefined") << endl;
cout << "10/0 = " << (r2 ? to_string(*r2) : "undefined") << endl;
```

```c++
// std::variant
using Number = variant<int, double, string>;

vector<Number> numbers = {42, 3.14, string("forty-two"), 100, 2.718};

for (const auto& n : numbers) {
    visit([](const auto& val) {
        cout << val << " (" << typeid(val).name() << ")" << endl;
    }, n);
}
```
