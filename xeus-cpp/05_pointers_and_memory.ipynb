# C Pointers & Memory Management

#### A pointer holds the address where another variable is stored in memory

Topics:

```c++
#include <stdio.h>
#include <stdlib.h>
#include <string.h>
```

<br>

## 1. Pointer Basics

```c++
int num = 42;
int* ptr = &num;  // ptr holds the address of num

printf("Value of num:     %d\n", num);
printf("Address of num:   %p\n", (void*)&num);
printf("Value of ptr:     %p\n", (void*)ptr);
printf("Address of ptr:   %p\n\n", (void*)&ptr);
printf("Dereferenced ptr: %d\n\n", *ptr);

// Modify via pointer
*ptr = 100;
printf("num after *ptr = 100: %d\n", num);
```

<br> 

## 2. Pointer Arithmetic

```c++
int arr[] = {10, 20, 30, 40, 50};
int* p = arr;  // Points to arr[0]

for (int i = 0; i < 5; i++) {
    printf("p+%d = %p, *(p+%d) = %d\n", i, (void*)(p+i), i, *(p+i));
}
// Pointer difference
printf("Difference between end and start: %td elements\n", (p+5) - p);
```

<br>

## 3. Pointers and Arrays

An array name decays to a pointer to its first element. `arr[i]` is equivalent to `*(arr + i)`.

```c++
int data[] = {1, 4, 9, 16, 25};
int* dp = data;

// Equivalent access methods
for (int i = 0; i < 5; i++) {
    printf("data[%d]=%d  *(data+%d)=%d  dp[%d]=%d  *(dp+%d)=%d\n",
           i, data[i], i, *(data+i), i, dp[i], i, *(dp+i));
}
```

```c++
// Strings are just char arrays — iterate with pointer
const char* str = "Lorem ipsum";
printf("Characters: ");
for (const char* cp = str; *cp != '\0'; cp++) {
    printf("%c ", *cp);
}
printf("\n");
```

## 4. Pointers to Functions

```c++
int add_fn(int a, int b) { return a + b; }
int sub_fn(int a, int b) { return a - b; }
int mul_fn(int a, int b) { return a * b; }

// Array of function pointers
int (*ops[3])(int, int) = {add_fn, sub_fn, mul_fn};
const char* names[] = {"add", "sub", "mul"};

int x = 10, y = 3;
for (int i = 0; i < 3; i++) {
    printf("%s(%d, %d) = %d\n", names[i], x, y, ops[i](x, y));
}
```

## 5. Dynamic Memory Allocation

#### `malloc`&nbsp;&nbsp;-&nbsp;&nbsp;allocate uninitialized memory

```c++
int n = 5;
int* heap_arr = (int*)malloc(n * sizeof(int));

if (heap_arr == NULL) {
    fprintf(stderr, "malloc failed!\n");
} else {
    for (int i = 0; i < n; i++) heap_arr[i] = (i + 1) * (i + 1);
    printf("Heap array: ");
    for (int i = 0; i < n; i++) printf("%d ", heap_arr[i]);
    printf("\n");
    free(heap_arr);  // Always free
    heap_arr = NULL; // Prevent dangling pointer
}
```

<br>

#### `calloc`&nbsp;&nbsp;-&nbsp;&nbsp;allocate zero-initialized memory

#### `realloc`&nbsp;&nbsp;-&nbsp;&nbsp;resize

```c++
int* zeros = (int*)calloc(6, sizeof(int));
printf("calloc (should be all 0s): ");
for (int i = 0; i < 6; i++) printf("%d ", zeros[i]);
printf("\n");

zeros = (int*)realloc(zeros, 10 * sizeof(int));
for (int i = 6; i < 10; i++) zeros[i] = i;
printf("After realloc (10 elements): ");
for (int i = 0; i < 10; i++) printf("%d ", zeros[i]);
printf("\n");

free(zeros);
zeros = NULL;
```

## 6. Common Memory Pitfalls


#### A pointer that points to freed memory - accessing it is undefined behavior.

```c++
// NB! Do not dereference dangling pointers in real code
int* dp_ex = (int*)malloc(sizeof(int));
*dp_ex = 55;
printf("Before free: %d\n", *dp_ex);
free(dp_ex);
dp_ex = NULL;  // Set to NULL immediately after free
// *dp_ex would be undefined behavior
```

<br>

## 7. General Memory Rules

- Always check malloc/calloc for NULL 
- Every malloc must have a matching free 
- Set pointer to NULL after free 
- Never free the same pointer twice 
- Never access freed memory
- Don't use uninitialized pointers
