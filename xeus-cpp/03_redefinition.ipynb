# Testing redefinition behaviour


#### In a traditional C++ workflow, you write code in a `.cpp` file, compile it all at once, and run it. If you want to change a variable type, you simply edit the file and recompile. However, **xeus-cpp**'s cell-based workflow enables several workarounds
---

In this example, we declare a variable with an `int` type, which we will name `a`, and assign a value to it

```c++
#include <iostream>

int a = 100;
std::cout << "Initial declaration: " << a << std::endl;
```

Next, we try to redeclare the variable, this time to `double`, and assign a new value to it

```c++
double a = 55.5;
std::cout << "New declaration: " << a << std::endl;
```

This action of redefinition is **illegal** in standard C++, because the variable `a` already exists as an `int`.

Two workarounds exist:
- overwriting the value **without redeclaring the type**
- creating a temporary "bubble" by redeclaring it in a **different scope**, so the variables don't interfere. This method doesn't change the value or the type of the global variable
  
***NB!*** It is advised that unless you absolutely must use the same name for a variable, you simply pick a new name

---

### First method - overwriting the value

```c++
a = 20.8;
std::cout << "Updated value: " << a << std::endl;
```

When using the first method to assign a floating point value to an `int`, the new value is floored to the closest integer

As shown in this example 

> 20.8 --> 20

### Second method - redeclaring in a different scope

```c++
{
    float a = 12.34f; //safe because it's inside a local scope
    std::cout << "Updated value (scoped): " << a << std::endl;
}

//Outside the brackets, the original int still exists
std::cout << "Global value: " << a << std::endl;
```
