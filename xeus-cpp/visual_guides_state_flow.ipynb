## 3. Memory allocation



````c++
#include <iostream>

int* ptr = new int(42);

std::cout << "Stack Frame: [ptr] variable address = " << (void*)&ptr << std::endl;
std::cout << "Heap Alloc:  Address pointed to     = " << (void*)ptr << std::endl;
std::cout << "Heap Value:  Value stored in heap   = " << *ptr << std::endl;
````

````c++
delete ptr;

std::cout << "Heap: Memory block deallocated." << std::endl;

std::cout << "\nDANGLING POINTER THREAT DETECTED" << std::endl;
std::cout << "Stack Frame: [ptr] still exists at address: " << (void*)&ptr << std::endl;
std::cout << "Danger zone: [ptr] still points to address: " << (void*)ptr << std::endl;
std::cout << "Undefined Behavior (Reading freed heap):     " << *ptr << " (Garbage or old value)" << std::endl;
````

````c++
ptr = nullptr;

std::cout << "Stack Frame: [ptr] variable address remains = " << (void*)&ptr << std::endl;
std::cout << "Stack State: [ptr] value is now safely set to = " << (void*)ptr << " (NULL)" << std::endl;

if (ptr == nullptr) {
    std::cout << "\nSuccess: Pointer is null" << std::endl;
}
````
