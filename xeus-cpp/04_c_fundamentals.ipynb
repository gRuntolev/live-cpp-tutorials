## 2. Operators and Expressions


#### Arithmetic operators

```c++
int a = 17, b = 5;
printf("%d + %d = %d\n", a, b, a + b);
printf("%d - %d = %d\n", a, b, a - b);
printf("%d * %d = %d\n", a, b, a * b);
printf("%d / %d = %d  (integer division)\n", a, b, a / b);
printf("%d %% %d = %d  (modulus)\n", a, b, a % b);
```

<br>

#### Bitwise operators

```c++
unsigned int x = 0b1010, y = 0b1100;
printf("x & y  =  %u  (AND)\n",  x & y);
printf("x | y  =  %u  (OR)\n",   x | y);
printf("x ^ y  =  %u  (XOR)\n",  x ^ y);
printf("~x     =  %u  (NOT)\n",  ~x);
printf("x << 1 =  %u  (left shift)\n", x << 1);
printf("x >> 1 =  %u  (right shift)\n", x >> 1);
```

## 3. Control Flow (`if`-`else`, `for` and `while` loops, `switch`

#### `if-else` condition

```c++
int score = 78;
if (score >= 90) {
    printf("Grade: A\n");
} else if (score >= 75) {
    printf("Grade: B\n");
} else if (score >= 60) {
    printf("Grade: C\n");
} else {
    printf("Grade: F\n");
}
```

<br>

#### `for` loop

```c++
// print multiplication table for 3
for (int i = 1; i <= 10; i++) {
    printf("3 x %2d = %2d\n", i, 3 * i);
}
```

<br>

#### `while` loop (demonstrated with a Collatz sequence)

```c++
int n = 7;
int steps = 0;
printf("Collatz sequence starting at %d: %d", n, n);
while (n != 1) {

    if (n % 2 == 0) //n is even
    {
        n /= 2;
    }
    else //n is odd
    {
        n = 3 * n + 1;
    }

    printf(" -> %d", n);

    steps++;
}
printf("\nTotal steps: %d\n", steps);
```

<br>

#### `switch` statement

```c++
char grade = 'B';
switch (grade) {
    case 'A': printf("Excellent!\n"); break;
    case 'B': printf("Good job!\n"); break;
    case 'C': printf("Average\n"); break;
    default:  printf("Needs improvement\n");
}
```

## 4. Functions

```c++
// Basic functions
int factorial(int n) {
    if (n <= 1) return 1;
    return n * factorial(n - 1);
}

int fibonacci(int n) {
    if (n <= 1) return n;
    return fibonacci(n - 1) + fibonacci(n - 2);
}

for (int i = 0; i <= 10; i++) {
    printf("%d! = %d\n", i, factorial(i));
}
```

<br>

```c++
printf("Fibonacci sequence: ");
for (int i = 0; i < 12; i++) {
    printf("%d ", fibonacci(i));
}
printf("\n");
```

## 5. Arrays and Strings

```c++
#include <string.h>

// Integer array
int nums[] = {5, 3, 8, 1, 9, 2, 7, 4, 6};
int len = sizeof(nums) / sizeof(nums[0]);

// Find min and max
int minV = nums[0], maxV = nums[0];
for (int i = 1; i < len; i++) {
    if (nums[i] < minV) minV = nums[i];
    if (nums[i] > maxV) maxV = nums[i];
}
printf("Array length: %d, Min: %d, Max: %d\n", len, minV, maxV);
```

<br>

```c++
// C strings
char name[] = "John";
char greeting[50];

strcpy(greeting, "Hello, ");
strcat(greeting, name);
strcat(greeting, "!");

printf("%s\n", greeting);
printf("Length of greeting: %zu\n", strlen(greeting));
```

## 6. Structs

```c++
struct Student {
    char name[50];
    int age;
    double grade;
};

void printStudent(struct Student s) {
    printf("Name: %s,  Age: %d,  Grade: %.2f\n", s.name, s.age, s.grade);
}
```

```c++
struct Student students[] = {
    {"Alice", 16, 5.78},
    {"Bob", 17, 5.42},
    {"Connor", 16, 4.96}
};

int numStudents = sizeof(students) / sizeof(students[0]);
for (int i = 0; i < numStudents; i++) {
    printStudent(students[i]);
}
```
